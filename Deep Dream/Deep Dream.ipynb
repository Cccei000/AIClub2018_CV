```python
import torch as t
import numpy as np
from torchvision import transforms
from PIL import Image
import utils
import modelnet
import Dreamer
```

```python
transform = transforms.Compose([
    transforms.ToTensor(),
    transforms.Normalize([0.485, 0.456, 0.406], [0.229, 0.224, 0.225])
])
input_img = Image.open('./starry_night.jpg')
input_tensor = transform(input_img)
input_np = input_tensor.numpy()
input_img
```

```python
model = modelnet.res50(True)
```

```python
Dreamer.dreamer(model,input_np,zoom_n = 6,zoom_ratio = 0.7,train = False,
                epochs = 15,max_jitter = 20,trg_layer = 3,lr = 0.02)
```
